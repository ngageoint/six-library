```python
import itertools
import h5py
from pathlib import Path
from datetime import datetime

import numpy as np
```

# Parameters

```python
NUM_PTS = 400
bin1 = ['foo', 'bar']
bin2 = ['dog', 'cat']
bin3 = ['a', 'b', 'c', 'd']
```

# Creates nested dictionary

```python
file_dict = {}
dummy_arr_real = np.ones((NUM_PTS,), dtype=np.float32)
dummy_arr_imag = np.zeros((NUM_PTS,), dtype=np.float32)

for idx in range(5):
    for jdx in range(len(bin1)):
        for kdx in range(len(bin2)):
            for ldx in range(len(bin3)):
                key = f'{idx + 1}/{bin1[jdx]}/{bin2[kdx]}/{bin3[ldx]}'
                file_dict[key + '/r'] = dummy_arr_real
                file_dict[key + '/i'] = dummy_arr_imag
print(file_dict)
```

# Creates H5 file

```python
out_path = 'nested_complex_float32_data_large.h5'
out_path
```

```python
def add_data(key, data_dict, handle):
    data = handle.create_group(key)
    for key, val in data_dict.items():
        grp = data
        sub_keys = key.split('/')
        for sub_key in sub_keys:
            if sub_key == sub_keys[-1]:
                grp.create_dataset(
                    sub_key,
                    data=val,
                    dtype=val.dtype
                )
                continue
            if sub_key not in grp:
                grp.create_group(sub_key)
            grp = grp[sub_key]
```

```python
with h5py.File(out_path, 'w') as handle:
    add_data('Data', file_dict, handle)
```
